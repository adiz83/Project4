# Merging file 1 and file 2

```python
import pandas as pd
```

```python
df1=pd.read_csv("file1.csv")
df2=pd.read_csv("file_2.csv")
```

```python
print(df1)
```

```python
df2['Average Number Of Words Per Sentence'] = df2['Average Sentence Length'] 
print(df2.columns)
```

```python
df1
```

```python
column_mapping_df1 = {
    'File': 'URL_ID',
    'Positive Score': 'POSITIVE SCORE',
    'Negative Score': 'NEGATIVE SCORE',
    'Polarity Score': 'POLARITY SCORE',
    'Subjectivity Score': 'SUBJECTIVITY SCORE',
}
```

```python
df1 = df1.rename(columns=column_mapping_df1) 
```

```python
column_mapping_df2 = {
    'File': 'URL_ID',
    'Average Sentence Length': 'AVG SENTENCE LENGTH',
    'Percentage of Complex Words': 'PERCENTAGE OF COMPLEX WORDS',
    'Fog Index': 'FOG INDEX',
    'Count of Complex Words': 'COMPLEX WORD COUNT',
    'Syllable Count Per Word': 'SYLLABLE PER WORD',
    'Personal Pronoun Counts': 'PERSONAL PRONOUNS',
    'Average Word Length': 'AVG WORD LENGTH',
    'Cleaned Word Count': 'WORD COUNT',
    'Average Number Of Words Per Sentence': 'AVG NUMBER OF WORDS PER SENTENCE',
}
```

```python
df2 = df2.rename(columns=column_mapping_df2)
print(df2)
```

```python
print(df1)
print(df2)
```

```python
merged_df = pd.merge(df1, df2, on='URL_ID', how='inner')

desired_columns_order = ['URL_ID', 'POSITIVE SCORE', 'NEGATIVE SCORE', 'POLARITY SCORE', 
                          'SUBJECTIVITY SCORE', 'AVG SENTENCE LENGTH', 'PERCENTAGE OF COMPLEX WORDS',
                          'FOG INDEX', 'AVG NUMBER OF WORDS PER SENTENCE', 'COMPLEX WORD COUNT',
                          'WORD COUNT', 'SYLLABLE PER WORD', 'PERSONAL PRONOUNS', 'AVG WORD LENGTH']

merged_df = merged_df[desired_columns_order]
merged_df['URL_ID'] = merged_df['URL_ID'].str.replace('.txt', '')
merged_df['URL_ID'] = merged_df['URL_ID'].str.replace('blackassign_', '')
print(merged_df)
```

```python
input_df = pd.read_excel('Input.xlsx') 
print(input_df)
```

```python
final_merged_df = pd.merge(merged_df, input_df, left_on='URL_ID', right_on='URL_ID', how='inner')
desired_columns_order = ['URL_ID', 'URL', 'POSITIVE SCORE', 'NEGATIVE SCORE', 'POLARITY SCORE', 
                          'SUBJECTIVITY SCORE', 'AVG SENTENCE LENGTH', 'PERCENTAGE OF COMPLEX WORDS',
                          'FOG INDEX', 'AVG NUMBER OF WORDS PER SENTENCE', 'COMPLEX WORD COUNT',
                          'WORD COUNT', 'SYLLABLE PER WORD', 'PERSONAL PRONOUNS', 'AVG WORD LENGTH']


final_merged_df = final_merged_df[desired_columns_order]
```
